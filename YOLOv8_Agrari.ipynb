```
import glob
import pandas as pd
import numpy as np

for f in glob.glob("/content/train/labels/*.txt"):
  filename = f.split("/")[-1]
  try:
    label = pd.read_csv(f, delimiter=" ", names=["code", "x", "y", "w", "h"])
    label["code"] = np.full(len(label), 0)
  except:
    label = pd.DataFrame()
  
  output_path = "/content/train/labels/" + filename
  label.to_csv(output_path, sep=" ", index=False, header=False)

for f in glob.glob("/content/valid/labels/*.txt"):
  filename = f.split("/")[-1]
  try:
    label = pd.read_csv(f, delimiter=" ", names=["code", "x", "y", "w", "h"])
    label["code"] = np.full(len(label), 0)
  except:
    label = pd.DataFrame()
  
  output_path = "/content/valid/labels/" + filename
  label.to_csv(output_path, sep=" ", index=False, header=False)

for f in glob.glob("/content/test/labels/*.txt"):
  filename = f.split("/")[-1]
  try:
    label = pd.read_csv(f, delimiter=" ", names=["code", "x", "y", "w", "h"])
    label["code"] = np.full(len(label), 0)
  except:
    label = pd.DataFrame()
  
  output_path = "/content/test/labels/" + filename
  label.to_csv(output_path, sep=" ", index=False, header=False)
```
